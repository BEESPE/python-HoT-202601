## <font color="green">9- Number Search by Dichotomy</font>

Binary search ([dichotomy](https://fr.wikipedia.org/wiki/Dichotomie)) is a method that divides a problem (usually in equal parts) to find the solution. For example, here is a discussion between Pierre and Patrick in which Pierre tries to guess the number (between 1 and 100) Patrick is thinking of.

- [Patrick] “I thought of a number between 1 and 100.”
- [Pierre] “OK, I will try to guess it. Is your number smaller or larger than 50?”
- [Patrick] “Larger.”
- [Pierre] “Is your number smaller, larger, or equal to 75?”
- [Patrick] “Larger.”
- [Pierre] “Is your number smaller, larger, or equal to 87?”
- [Patrick] “Smaller.”
- [Pierre] “Is your number smaller, larger, or equal to 81?”
- [Patrick] “Smaller.”
- [Pierre] “Is your number smaller, larger, or equal to 78?”
- [Patrick] “Larger.”
- [Pierre] “Is your number smaller, larger, or equal to 79?”
- [Patrick] “Equal. That’s the number I thought of. Congrats!”

To quickly guess the number, the trick is to take the midpoint of the interval containing the number at each step. Steps:

- Number is between 1 and 100, propose 50 (100 / 2).
- Number is between 50 and 100, propose 75 (50 + (100 - 50)/2).
- Number is between 75 and 100, propose 87 (75 + (100 - 75)/2).
- Number is between 75 and 87, propose 81 (75 + (87 - 75)/2).
- Number is between 75 and 81, propose 78 (75 + (81 - 75)/2).
- Number is between 78 and 81, propose 79 (78 + (81 - 78)/2).
